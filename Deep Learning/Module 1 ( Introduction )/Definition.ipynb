Artificial Intelligence: Artificial Intelligence is the broader field of computer science that focuses on building systems capable of performing tasks that normally require human intelligence — such as reasoning, learning, decision-making, understanding language, and perception.
Machine Learning and Deep Learning are subsets of Artificial Intelligence.


Machine Learning: Machine Learning is a subset of AI that allows systems to learn patterns from data and improve performance over time without being explicitly programmed.
Types of ML:

Supervised Learning
Data has labels (input → output known)
Example: House price prediction

Unsupervised Learning
No labels, finding patterns
Example: Customer segmentation

Reinforcement Learning
Learning by reward and punishment
Example: Game-playing AI


Deep Learning: 
Deep Learning is a subset of Machine Learning that uses artificial neural networks with multiple layers to automatically learn complex patterns from large amounts of data.
Real Life usage: face recognition
Self Driving Car
Chatgpt


History of AI:
Perceptron: Perceptron is the building block of neural networks. It performs binary classification by learning a linear decision boundary.

Learning mechanism of Neural Network :
A neural network is a layered system of connected neurons that learns patterns from data to make predictions.
