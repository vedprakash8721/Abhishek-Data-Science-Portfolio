```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x,y=make_moons(n_samples=1000,noise=0.2,random_state=42)
y=y.reshape(-1,1) #1D array ko 2D column vector me convert karna
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)

# Building sequential model with linear activation function
def build_linear_model():
    model=Sequential([
        Dense(4,activation='linear',input_dim=2), # ReLU = Rectified Linear Unit
        Dense(1,activation='sigmoid',) # output layer for binary classification
    ])
    model.compile(optimizer=Adam(learning_rate=0.2),loss='binary_crossentropy',metrics=['accuracy'])
    return model

# Building sequential model with relu activation function
def buil_relu_model():
    model=Sequential([
        Dense(4,activation='relu',input_dim=2),
        Dense(1,activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.1),loss='binary_crossentropy',metrics=['accuracy'])
    return model

# train and evaluate model
def train_and_evaluate(model,x_train,y_train,x_test,y_test,epochs=50,batch_size=35):
    model.fit(x_train,y_train,epochs=epochs,batch_size=batch_size,verbose=0)
    y_pred=(model.predict(x_test)>0.5).astype(int)
    accuracy=accuracy_score(y_test,y_pred)
    return accuracy,y_pred

# Initializing Model
linear_model=build_linear_model()
relu_model=buil_relu_model()

# Train and evaluate linear model
print("Training linear model ..................")
linear_accuracy,linear_pred=train_and_evaluate(linear_model,x_train,y_train,x_test,y_test)
print(f"accuracy with linear activation function: {linear_accuracy}")

# Train and evaluate with relu model
print("Training relu model...................")
relu_accuracy,relu_pred=train_and_evaluate(relu_model,x_train,y_train,x_test,y_test)
print(f"accuracy with relu activation function :{relu_accuracy}")
```
