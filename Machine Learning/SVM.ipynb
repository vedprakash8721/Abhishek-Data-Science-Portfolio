```python
""" importing all the library so that in future it """
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn import datasets
import plotly as px
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
'''

SVM : Support Vector Machine (SVM) is a supervised machine learning algorithm mainly used for classification, sometimes for regression.
It tries to find the best boundary (hyperplane) that separates classes in your dataset.

''' 
```

```python
iris=datasets.load_iris()
# description about this data set
""" 
load_iris is a dataset about the flowers ,
sepal:
The outer greenish part of the flower, like small leaves that protect the bud before it blooms.
Petal:
The colored part of the flower (the "beautiful" part we usually notice).
"""
x=iris.data
y=iris.target
names=iris.target_names
display(x.shape)
display(y.shape)
df=pd.DataFrame(x,columns=iris.feature_names)
df['species']=iris.target
df['species']=df['species'].replace(to_replace=[0,1,2],value=['rose','sunflower','tulips'])
display(df)
sns.pairplot(data=df,hue='species',palette='Set2')
# fig=px.scatter(df,x='petal length (cm)',y='petal width (cm)',color='species')
# fig.show()
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
display(x_train.shape)
display(x_test.shape)
svm=SVC(kernel='linear',random_state=0) # random_state=ensures results are reproducible. Every time you run the code, you get the same behavior
svm.fit(x_train,y_train)
pred=svm.predict(x_test)
display(pred)
acc=accuracy_score(y_test,pred)
print(f"The Accuracy of the Model is : {acc*100} %")
```
