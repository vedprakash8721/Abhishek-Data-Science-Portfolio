```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# sometimes when you use some fun are removed in python to remove the warnings import
import warnings 
warnings.filterwarnings('ignore') # suppress all the warnings
from IPython.display import display
```

```python
# lets start the session
mt_df=pd.DataFrame(np.random.randint(50,100,100),columns=['Data'],index=pd.date_range('2023-9-6',periods=100))
display(mt_df)

# our first plot 
mt_df.plot(figsize=(20,5),
           color='green',
           marker='o',
           linestyle='--',
           # linewidth=1
           )
plt.title("Line chart")
 # we can decorate the Graph by using color and so on 
# in ( x=length, y = width) of the Graph
plt.show()
""" Scatter Plot: it is a type of visualization in which there is two data in the axis  ( cartesian planes ) """
x=np.random.randint(60,120,200)
y=np.random.randint(70,120,200)
# plotting the Cartesian graph using matplotlib
plt.scatter(x,y,color='blue')
# plt.grid()
plt.xlabel("X values")
plt.ylabel(" Y values")
plt.title(" Scatter Plot Example")
plt.show()

# Multiple line in one Graph
x1=np.sort(np.random.choice(np.arange(1,20),size=10,replace=False))
y1=np.sort(np.random.choice(np.arange(10,60),size=10,replace=False))
y2=np.sort(np.random.choice(np.arange(15,35),size=10,replace=False))
plt.plot(x1,y1,color='r',marker='o',linestyle='--',label='Suzlon Engery')
plt.plot(x1,y2,color='b',marker='o',linestyle='--',label='JP Power')
plt.title(" Multiple Stock in one Graph")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()
print(x1)
```
