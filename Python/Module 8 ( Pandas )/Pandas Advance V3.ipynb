```python
""" How to merge two data frames : """
# Pandas Advance V3: 
# generating random data for data frame
import numpy as np
import pandas as pd
data1={
    "key1": [1,45,2,7,44,76,4,9],
    "key2": np.random.randint(1,100,8),
    "key3": np.random.randint(1,100,8)
}
data2={
    "key1": [1,3,4,6,3,5,2,9],
    "key4": np.random.randint(1,100,8),
    "key5": np.random.randint(1,100,8)
}

# data2={
#     "key1": np.random.randint(1,90,8),
#     "key4": np.random.randint(1,100,8),
#     "key5": np.random.randint(1,70,8)
# }

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
from IPython.display import display
display("this is data frame 1",df1)
print("-------------------------Breaking the line----------------------------\n")
display("this is data frame 2",df2)


# Merge: inner merge ( the common elements will be merged )
display(pd.merge(df1,df2, how='inner')) #  the common element like from key 1 from both df will be print 
print("-------------------------Breaking the line----------------------------\n")
display(pd.merge(df1,df2,how='left')) #all elements of df1 and common elements of key 1 from both will be print 
print("-------------------------Breaking the line----------------------------\n")
display(pd.merge(df1,df2,how='right'))
print("-------------------------Breaking the line----------------------------\n")


# outer merge:
display(pd.merge(df1,df2,how='outer')) # all elements of key1 from both data frame will be present and common elements from key1 will be print

print("-------------------------Breaking the line----------------------------\n")
print(pd.merge(df1,df2,how="cross")) # all the elements will come like first row of key 1 from df1 will merged with all rows of df2 and so on
```

```python
# some advance merge in data frames
display(pd.merge(df1,df2,how='left',left_on="key1",right_on="key4")) # as you can see how="left" then all elements of df1 and according to the condition leff on and right on the common from them will be print
```

```python
# Pandas Advance V3: 
import numpy as np
# Join
""" difference between join and merge: join works on index and merge works on column
"""
df1=pd.DataFrame({
    "key1": np.random.randint(1,10,6),
    "key2": np.random.randint(1,10,6),
    "key3": np.random.randint(1,10,6)
},index=["a","b","c","d","e","f"]
)

df2=pd.DataFrame({
    "key4": np.random.randint(1,10,6),
    "key5": np.random.randint(1,10,6),
    "key6": np.random.randint(1,10,6)
},index=["a","b","i","d","j","k"]
)

print("-------------------------Breaking the line----------------------------\n")
display(df1)
print("-------------------------Breaking the line----------------------------\n")
display(df2)
print("-------------------------Breaking the line----------------------------\n")

# lets perfom join : inner join
display(df1.join(df2,how="inner")) # this would be syntax :
print("-------------------------Breaking the line----------------------------\n")

# Join: left
display(df1.join(df2,how="left")) # all element of left df and common element from df2 on the basis of index
print("-------------------------Breaking the line----------------------------\n")

# and so on ............. 
```
