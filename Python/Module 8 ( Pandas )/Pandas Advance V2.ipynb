```python
import pandas as pd
data_frame=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
# some operations
print(data_frame.head(3))
print(data_frame.shape)

```

```python
from IPython.display import display
# access the rows and columns of the data frames 
""" use of loc and iloc"""
# iloc is used for integer indexing and loc is used for name
# if you want to access the frst n rows
display(data_frame.iloc[0:2]) # [ start: end-1]
""" and loc is used for names. loc is also known as location. in this the index value is considered as name """
display(data_frame.loc[0:3])
# in the output the indexing value 0,1,2,3 is considered as names
```

```python
# use of loc : if you want to print the first n rows but with specific columns
from IPython.display import display
display(data_frame.loc[0:30,["Name","Sex","Fare"]])
```

```python
# use of iloc : iloc always requires python internal index
# Pandas Advance V2 : 
display(data_frame.iloc[0:6,0:4])
```

```python
# to drop some column from the dataframe
# data_frame.drop("PassengerId")  python detect it as row. to specify column we have to give axis=column_num
display(data_frame.drop("PassengerId",axis=1)) 
""" axis indexing starts from 0 ( by default)"""
# the modification will not be shown if you print again the dataframe . to keep it permanent you have to assign it
""" when i try to delete the row 1 it raise key error"""
print("-------------------------Breaking the line----------------------------\n")

# to reset the action
data_frame.reset_index(drop=True)

# to set a index as any column
display(data_frame.set_index("Name"))

print("-------------------------Breaking the line----------------------------\n")

# Access the specific detail of the passengers
display(data_frame[data_frame["Name"]=="Heikkinen, Miss. Laina"])


```
