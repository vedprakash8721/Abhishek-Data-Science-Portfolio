```python
num=int(input("enter the num"))
def check_prime(num):
    if num==1 or num==2:
        print("num is prime")
        return
    
    if num>2:
        for i in range(2,num):
            if num%i==0:
                print("no is not prime")
                return
        print("no is prime")
check_prime(num)
```
