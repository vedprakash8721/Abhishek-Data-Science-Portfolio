```python
import duckdb
import pandas as pd
from IPython.display import display
df=pd.read_csv(r"C:\MY SPACE\Data Science and Gen AI\Abhishek-Data-Science-Portfolio\SQL\SQL PRACTICE\E-COMMERCE\data\flipkart.csv")
conn=duckdb.connect()
conn.register("flipkart",df)
result=conn.execute("select country from flipkart where age>=50 and gender='Female' group by country").df()
print(result)
display(df.head(5))
```

```python
result=conn.execute("select country from flipkart where age>=50 and gender='Male' group by country").df()
print(result)
```

```python
result = conn.execute("""select country, gender, count(1) from flipkart group by 1,2
 """).df()
display(result.head(10))
```

```python
result=conn.execute(""" select country, count(*) as male_count from flipkart
                    where gender='Male'
                    group by country
                    order by male_count desc
""").df()
display(result)
```

```python
result=conn.execute(""" select country,gender, count(*) as count from flipkart
                    where gender in ('Male','Female','Other')
                    group by country,gender
                    order by country, count desc
""").df()
display(result)
```

```python
result=conn.execute(""" select
             sum(case when gender='Male' then 1 else 0 end) as male_count,
             sum(case when gender='Female' then 1 else 0 end) as female_count,
             sum(case when gender='Other' then 1 else 0 end) as other_count
            from flipkart where gender in ( 'Male','Female','Other')
""").df()
display(result)
```
