```python
import pandas as pd 
import duckdb

df=pd.read_csv(r'c:/MY SPACE/DATA SCIENCE AND GEN AI/Abhishek-Data-Science-Portfolio/SQL/SQL PRACTICE/E-COMMERCE/data/flipkart.csv')
conn=duckdb.connect()
conn.register("flipkart",df)
```

```python
result=conn.execute(""" select country from flipkart
        where age>=50  and gender='Female' group by country
""").df()
print(result)
```
