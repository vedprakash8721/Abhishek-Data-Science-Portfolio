```python
# using operator with where clause:
""" 
arithmatic op ( =,-,*,/)
comparison op ( =, !=,<,> etc)
logical op ( and, or,not,between)
bitwise op 
"""
# logical and me dono condition true honi chahiye
result=conn.execute("""select name,marks
                    from student
                    where marks>80 and city='Lucknow'
""").df()
display(result)
result=conn.execute("""select name,marks
                    from student
                    where marks>80 or city='Lucknow'
""").df()
display(result)

result=conn.execute("""select * 
                    from student 
                    order by marks desc
                    limit 3""").df()
display(result)
```

```python
# AGGREGATE FUCNTION:

# MIN AND MAX
result=conn.execute("select max(marks) from student").df()
display(result)
result=conn.execute(""" select name,marks
                    from student
                    order by marks desc 
                    limit 1""").df()
display(result)

# Group by clause:
result=conn.execute(""" select city, count(name)
                    from student
                    group by 1
                    order by count(name) desc""").df()
display(result)
```

```python
result=conn.execute(""" select mode_of_payment,count(customer_name)
                    from payments
                    group by mode_of_payment
                    order by count(customer_name) desc
""").df()
display(result)

result=conn.execute("""select
                    sum(case when mode_of_payment='UPI' then 1 else 0 end) as UPI,
                    sum(case when mode_of_payment='Debit Card' then 1 else 0 end) as Debit_Card,
                    sum(case when mode_of_payment='Credit Card' then 1 else 0 end) as Credit_Card,
                    sum(case when mode_of_payment='Net Banking' then 1 else 0 end) as Net_Banking
                    from payments where mode_of_payment in ('UPI','Debit Card','Net Banking','Credit Card')
""").df()
display(result)
```
