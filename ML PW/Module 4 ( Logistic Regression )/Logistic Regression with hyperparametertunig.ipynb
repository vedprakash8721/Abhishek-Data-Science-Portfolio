```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from IPython.display import display
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_redundant=5,n_informative=5,random_state=42,n_classes=2)
# n_informative: Model learns mainly from these
# n_redundant= Features that are derived from informative features If f1 and f2 are informative → redundant could be f3 = f1 + f2
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
param={'penalty':('l1','l2','elasticnet'),'C':[1,10,20,30,40]}
classifier=LogisticRegression()
clf=GridSearchCV(classifier,param_grid=param,verbose=2)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)
display(acc)
clf.best_params_
```
