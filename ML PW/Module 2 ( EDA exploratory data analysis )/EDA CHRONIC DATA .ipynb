```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
```

```python
# ------------------------------- CONVERTING OBJECT TO INT/FLOAT AS COLUMN SHOULD BE -------------------------------
df=pd.read_csv('kidney_disease.csv')
display(df.head(5))
display(df.columns)
df.drop('id',axis=1,inplace=True)
df.columns=[
'Age (years)',
'Blood Pressure (mm/Hg)',
'Specific Gravity (urine)',
'Albumin (urine)',
'Sugar (urine)',
'Red Blood Cells (urine)',
'Pus Cell (urine)',
'Pus Cell Clumps',
'Bacteria (urine)',
'Blood Glucose Random (mg/dL)',
'Blood Urea (mg/dL)',
'Serum Creatinine (mg/dL)',
'Sodium (mEq/L)',
'Potassium (mEq/L)',
'Hemoglobin (g/dL)',
'Packed Cell Volume (%)',
'White Blood Cell Count (cells/cumm)',
'Red Blood Cell Count (millions/cumm)',
'Hypertension',
'Diabetes Mellitus',
'Coronary Artery Disease',
'Appetite',
'Pedal Edema',
'Anemia',
'Chronic Kidney Disease Classification'
]
display(df.columns)
display(df.head(5))
display(df['Serum Creatinine (mg/dL)'].sort_values(ascending=False)[:10])
display(df.shape)
display(df.describe())
display(df.info())
display(df['Red Blood Cells (urine)'])
display(df['Pus Cell (urine)'])
display(df['Bacteria (urine)'])
display(df['Packed Cell Volume (%)'].unique())
df['Packed Cell Volume (%)']=pd.to_numeric(df['Packed Cell Volume (%)'],errors='coerce')
display(df['Packed Cell Volume (%)'].dtypes)
df['White Blood Cell Count (cells/cumm)']=pd.to_numeric(df['White Blood Cell Count (cells/cumm)'],errors='coerce')
df['Red Blood Cell Count (millions/cumm)']=pd.to_numeric(df['Red Blood Cell Count (millions/cumm)'],errors='coerce')
```

```python
display(df.columns)
display(df['Chronic Kidney Disease Classification'].unique())
df['Chronic Kidney Disease Classification']=df['Chronic Kidney Disease Classification'].str.strip()
```

```python
# ------------------------------ VISUALIZATION ------------------------------------
plt.figure(figsize=(14,6))
sns.countplot(x='Chronic Kidney Disease Classification',data=df,palette={'ckd':'red','notckd':'green'})
plt.show()
```

```python
# separating the categorical and numerical column
cat_col=[col for col in df.columns if df[col].dtype=='object'] # this loop returns list value for cat_col not dataframe 
display(cat_col)
# to see the value in the columns 
for col in cat_col:
    print(f"{col} has {df[col].unique()} values ")
num_col=[col for col in df.columns if df[col].dtype!='object']
display(num_col)
for col in num_col:
    print(f"{col} has {df[col].unique()} values ")

# by using strip() converting the column into 
df['Diabetes Mellitus']=df['Diabetes Mellitus'].str.strip()
df['Coronary Artery Disease']=df['Coronary Artery Disease'].str.strip()
# Conveting the 'Chronic Kidney Disease Classification' column into int column like o for not ckd and 1 for ckd for better visualization 
df['Chronic Kidney Disease Classification']=df['Chronic Kidney Disease Classification'].map({'ckd':1,'notckd':0})
display(df.columns)
```

```python
# Visualizing the Age frequency for Kidney Disease
plt.figure(figsize=(16,6))
sns.countplot(x='Age (years)',data=df,hue='Age (years)',legend=True)
plt.xticks(range(0,90,5))
plt.yticks(range(0,20,2))
plt.xlabel('At Age patient has Kidney Disease')
plt.ylabel('Frequency(no of Patients)')
plt.title('visualizing at which age patients has ckd ')
plt.legend()
plt.show()

# Appetite
plt.figure(figsize=(10,8))
df['Appetite'].value_counts().plot.pie(autopct='%1.1f%%',explode=(0,0.09),shadow=True,startangle=80)
plt.title(" Appetite (Hunger) condition in patients ")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['Coronary Artery Disease'].value_counts().plot.pie(autopct='%1.1f%%',wedgeprops=dict(width=0.5),shadow=True,explode=(0,0.0))
plt.title('Coronary Artery Disease ( patient with heart issue)')
plt.legend()
plt.show()
```

```python
# Stack bar chart
st_chart=pd.crosstab(df['Diabetes Mellitus'],df['Hypertension'])
st_chart.plot(kind='bar',stacked=True,colormap='coolwarm',figsize=(16,6))
plt.xlabel('Diabetes Patient')
plt.ylabel('number of patient')
plt.title('Diabetes and Hypertension relationship in patients')
plt.yticks(range(0,300,20))
plt.legend(title='Hypertension')
```

```python
# Pair Plot
display(num_col)
plt.figure(figsize=(14,8))
cols=['Age (years)','Sugar (urine)','Blood Urea (mg/dL)','Serum Creatinine (mg/dL)','Chronic Kidney Disease Classification']
g=sns.PairGrid(df[cols],hue='Chronic Kidney Disease Classification',palette='coolwarm')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot,cmap='Blues_d')
g.map_diag(sns.histplot)
g.add_legend()
plt.show()
```
