OOB Score (Out-of-Bag Score) is an internal validation score used by Random Forest to estimate model performance without needing a separate validation dataset.

How it works:------------

Each tree is trained on a bootstrap sample (random sample with replacement).

About 63% of the training data is used to build a tree.

The remaining 37% (called Out-of-Bag samples) are not used for that tree.

These OOB samples are used to test the tree's predictions.

The combined accuracy from all OOB predictions is the OOB Score.


Why use it?---------------


No need for a separate validation set.

Saves data.

Gives a good estimate of the model's generalization performance.

```python
import warnings 
warnings.filterwarnings('ignore')
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model=RandomForestClassifier(oob_score=True)
model.fit(x_train,y_train)
model.oob_score_
```
