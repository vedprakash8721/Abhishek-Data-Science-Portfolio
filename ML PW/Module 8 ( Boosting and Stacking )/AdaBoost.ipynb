AdaBoost: Adaptive Boosting.
AdaBoost (Adaptive Boosting) is a boosting algorithm that combines multiple weak learners (usually decision stumps) to create a strong learner.

It works by giving more importance (higher weight) to the data points that previous models classified incorrectly.

Why is it called Adaptive?

Because it adapts after every iteration.

Correctly classified samples → weight decreases
Incorrectly classified samples → weight increases

decision tree stump: 

A decision stump is a decision tree with:
max_depth = 1

Why use Decision Stumps?

Because AdaBoost is designed to improve many simple models instead of relying on one complex tree.



```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from IPython.display import display
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
display(x)
display(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
# simple AdaBoost model:
clf=AdaBoostClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc_clf=accuracy_score(y_test,y_pred)
display(f"accuracy under standard adabosst: {acc_clf}")
```

```python
# With Hyperparameter tuning:
ada=AdaBoostClassifier(random_state=42)
param_grid={
    'n_estimators':[10,20,30,40,50],
    'learning_rate':[0.2,0.4,0.8,1]
}
model=GridSearchCV(estimator=ada,param_grid=param_grid,cv=5,scoring='accuracy',verbose=1)
model.fit(x_train,y_train)
y_pred_hype=model.predict(x_test)
acc_hype=accuracy_score(y_test,y_pred_hype)
print(f" accuracy under hyperparameter tuning model: = {acc_hype}")
print(f" classification reprot{classification_report(y_test,y_pred_hype)}")
print(f" confusion matrix {confusion_matrix(y_test,y_pred_hype)}")
print(model.best_params_)
```
