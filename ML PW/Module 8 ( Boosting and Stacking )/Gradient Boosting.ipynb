Gradient Boosting is a boosting ensemble algorithm where each new decision tree tries to reduce the errors (residuals) made by the previous trees using gradient descent.

Gradient Descent : Gradient Descent is an optimization algorithm used to find the best model parameters by minimizing the loss (error) function.

Loss function: A loss function measures the difference between the predicted value and the actual value.

Gradient Boosting is generally more powerful and flexible than AdaBoost because it minimizes residual errors using gradient descent, while AdaBoost mainly focuses on misclassified samples. 
Both can be used for classification and regression, but Gradient Boosting often gives better performance on complex datasets.


```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
```

```python
x,y=make_regression(n_samples=500,n_features=4,noise=10,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
Gb=GradientBoostingRegressor()
param_grid={
    'n_estimators':[30,40,50],
    'learning_rate':[0.1,0.2,0.3,0.5,1],
    'min_samples_split':[2,3,4],
    'min_samples_leaf':[1,2,3],
    'max_depth':[3,4,5]
}
model=GridSearchCV(estimator=Gb,param_grid=param_grid,cv=3,verbose=2,scoring='r2')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
display(r2)
```
