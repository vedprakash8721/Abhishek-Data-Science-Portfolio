# Naive Bayes: 

A supervised machine learning classification algorithm based on Bayes' Theorem. It predicts the class with the highest probability
P(A|B) = {P(B|A)P(A)}/P(B)


It assumes that all input features are independent of each other
Age and Salary are assumed independent while predicting whether a customer buys a product.
This assumption is usually not true, but the algorithm still works surprisingly well.

Variants of Naive Bayes:

Bernauli NB:if the feature follows the bernauli distribution then we use bernauli naive bayes classifier ( feature are / could be in value 0 and 1 )

Multinomial NB: when inputs are in text data : email classification : input-> email body -> ads ( discount ) this is spam / offer letter is ham ( not spam)

Mutlinomila and bernauli is used in nlp ( for text data )

Gausian Naive Bayes: if the feature are continuous and following normal distribution then we will use Gausian NB 

Advantages:
Very fast
Easy to implement
Works well on small datasets
Good for text classification
Handles high-dimensional data

Disadvantages:

Assumes feature independence
Doesn't perform well when features are highly correlated


it is only for classification problems

```python
# Implementation of Gausian Naive Bayes 
import numpy as np
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
from IPython.display import display
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
```

```python
x,y=load_iris(return_X_y=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size= 0.25,random_state=42)
# features are in continuous in nature do using Gaussian NB: 
cls=GaussianNB()
cls.fit(x_train,y_train)
y_pred=cls.predict(x_test)
ac=accuracy_score(y_test,y_pred)
display(ac)
```
