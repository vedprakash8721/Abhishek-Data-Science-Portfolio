```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error,mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from IPython.display import display
```

```python
df=pd.read_csv('height-weight.csv')
display(df.head(5))
display(df.shape)
display(df.info())
```

```python
x=df[['Weight']]
y=df['Height']
#display(x,y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
# Scaling : fit_transform is used on train data and transform is used on test data.
# fit is used to learn the pattern and test data is unseen data so if you use this you making a scene of data leakage 
scaler=StandardScaler()
"""
scaling should be used after train test split
target variable should not be scaled
avoid scaling categorical features
"""
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
display(x_train)
display(x_test)
LinearRegression()
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print("The slope or coef of the model is ",model.coef_)
print(" the intercept of the model is ", model.intercept_)
mae=mean_absolute_error(y_test,y_pred)
display(mae)
acc=mean_absolute_percentage_error(y_test,y_pred)
print(f"the accuracy of the model is {round(acc*100)}%")
```
