WHAT IS REGRESSION AND CLASSIFICATION ?

:- BOTH ARE SUPERVISED LEARNING TECHNIQUE
REGRESSION: REGRESSION IS ML TECHNINQUE THAT IS USED TO PREDICT NUMERCIAL VALUE.
REGRESSION ANSWERS : HOW MUCH , WHAT AND WHAT WOULD BE THE NUMBER.
USE IN REAL WORLD: HOUSE PREDICTION , TEMPRETURE, CRICKET SCORE, SALARY OF EMPLOYEE.

CLASSIFICATION: CLASSIFICATION IS ML TECHNIQUE THAT IS USED TO PREDICT CATEGORY OF CLASS.
IT ANSWERS IN TRUE FALSE , YES OR NO , 
REAL WORLD USE: MAIL IS SPAM OR NOT . DESEASE ( YES OR NOT ), TRANSACTION IS FRAUD OR LEGITIMATE.

GRADIENT DESCENT : GRADIENT DESCENT IS THE OPERATIONAL BACKBONE OF MODEL OPTIMIZATION . ITERATIVELY MINIMIZE THE LOSS FUNCTION BY MOVING PARAMETERS TO REACH LEAST ERROR VALUE.

MULTIPLE LINEAR REGRESSION ( MLR ) : MULTIPLE LINEAR REGRESSION IS A SUPERVISED LEARNING MODEL USED TO QUANTIFY THE LINEAR RELATIONSHIP BETWEEN ONE DEPENDENT VARIABLE AND MULTIPLE INDEPENDENT VARIABLES.
SLR:- ( ONE INDEPENDENT VARIABLE )
MLR:- ( MULTIPLE INDEPENDENT VARIABLE )

INERPRETATION OF COEFFICIENTS :  IN MULTIPLE LINEAR REGRESSION, EACH COEFFICIENT REPRESENTS THE EXPECTED CHANGE IN THE DEPENDENT VARIABLE FOR A ONE-UNIT CHANGE IN THE PREDICTOR, ASSUMING ALL OTHER VARIABLES ARE HELD CONSTANT.
