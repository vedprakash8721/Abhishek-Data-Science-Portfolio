```python
# Essential Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error,mean_absolute_percentage_error
```

```python
# Reading the data
from sklearn.datasets import load_diabetes
load_diabetes()
db=load_diabetes()
display(db.data)
db.target
db.feature_names
# Converting this into Dataframe
df=pd.DataFrame(db.data,columns=db.feature_names)
df['target']=db.target
display(df)
```

```python
# One independent variable 
x=df[['bmi']]
y=df['target']
# display(x,y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)# this splits the data randomly and if you want to fix the split use random_state=1
display(x_train)
LinearRegression()
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
plt.figure(figsize=(15,6))
plt.scatter(x_test,y_test,label='Actual')
plt.plot(x_test,y_pred,label='Predicted',color='r')
plt.xlabel('bmi')
plt.ylabel('Target')
plt.legend()
plt.show()

# Checking the accuracy of the model 
acc=mean_absolute_percentage_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
display(f" mean absolute error is {mae}")
display(f" mean squred error is {mse}")
display (f" The Accuracy of the Model is {acc*100}")
```
