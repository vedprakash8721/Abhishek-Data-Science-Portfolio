k-Nearest Neighbors (k-NN):

A supervised machine learning algorithm that predicts the class (classification) or value (regression) of a new data point by looking at the k closest data points (neighbors) in the training data.

Classification: Predicts the majority class of the k nearest neighbors. 
In classification, it predicts the class that best matches the data point based on the majority.

Regression: Predicts the average of the k nearest neighbors' values. 

Example:
If k = 3 and the nearest neighbors are Cat, Cat, Dog, the prediction is Cat.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
from IPython.display import display
display(x.shape)
display(x)
display(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_neighbors':[3,5,7,9,11],
    'weights':['uniform','distance'],
    'algorithm':['kd_tree','ball_tree','brute']
}
model=GridSearchCV(estimator=knn,param_grid=param_grid,cv=4,verbose=2)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
acc=accuracy_score(y_test,y_pred)
print(f" Accuracy under knn with hyptuning=  {acc}")
print(classification_report(y_test,y_pred))
```

```python
print(model.best_params_)
print(model.best_score_)
# predict using best_estimator
best_model=model.best_estimator_
y_pred_best=best_model.predict(x_test)
acc_best=accuracy_score(y_test,y_pred_best)
print(f"accuracy_score under best model{acc_best}")
```
