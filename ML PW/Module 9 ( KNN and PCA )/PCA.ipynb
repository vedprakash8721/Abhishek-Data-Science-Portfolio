# PCA : Principle Component analysis 

PCA is an unsupervised machine learning algorithm (or technique) used for dimensionality reduction.
It reduces the number of features while preserving as much variance as possible.
It is mainly used as a preprocessing step before training ML models.

Benefit: Faster training, less memory usage, reduced overfitting, and easier visualization.
Used for: High-dimensional datasets, image processing, preprocessing before ML models, and data visualization.

# curse of dimensionality:

After a certain point, adding more features provides little or no improvement in model performance,
while increasing training time, memory usage, and the risk of overfitting.
