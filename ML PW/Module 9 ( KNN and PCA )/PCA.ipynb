# PCA : Principle Component analysis 

PCA is an unsupervised machine learning algorithm (or technique) used for dimensionality reduction.
It reduces the number of features while preserving as much variance as possible.
It is mainly used as a preprocessing step before training ML models.

Benefit: Faster training, less memory usage, reduced overfitting, and easier visualization.
Used for: High-dimensional datasets, image processing, preprocessing before ML models, and data visualization.

# curse of dimensionality:

After a certain point, adding more features provides little or no improvement in model performance,
while increasing training time, memory usage, and the risk of overfitting.

# feature selection and extraction technique:

Principal Components are new features created by combining the original features.

```python
# Rituals
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=load_iris()
x=data.data
y=data.target
display(x)
display(x.shape)
display(y.shape)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
display(x_train)
```

```python
pca=PCA(n_components=3)
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)
display(x_train)
display(pca.explained_variance_ratio_)
```
