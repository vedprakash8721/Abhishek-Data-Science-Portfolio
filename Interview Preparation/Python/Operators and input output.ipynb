Q21. What are the different categories of operators in Python?

Python operators are special symbols used to perform operations on variables and values. The main categories are:

Arithmetic Operators
Comparison (Relational) Operators
Assignment Operators
Logical Operators
Bitwise Operators
Membership Operators
Identity Operators

---
Q22. Explain the difference between logical operators and bitwise operators

Logical operators work with Boolean values or conditions and return True or False. Bitwise operators work on the binary representation of integers, and returning an integer.

Q23. What is the difference between membership operators and identity operators? 

Membership operators check whether an element exists in a sequence or collection.
Identity operators check whether two variables refer to the same object in memory.

```python
# Q27. Write a Python program to calculate the area of a circle based on user input. 
import math
r=float(input(" enter the radius of the circle in cm "))
area=math.pi*r**2
print(f"the area of the circle is {round(area,4)} square cm")
```

```python
# Q28. Write a Python program to swap two numbers entered by the user without using a third variable. 

# method 1 : tuple unpacking:
a=int(input(" enter the first number "))
b=int(input("enter the second number "))
print(f"befor swapping: a= {a} and b = {b}")
a,b=b,a
print(f"after swapping a={a} and b={b}")

print('using addition method')
a=a+b
b=a-b
a=a-b
print(f"after swapping a={a} and b={b}")
```
