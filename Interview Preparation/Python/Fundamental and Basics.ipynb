Q1. What is Python, and what are its key features?

Python is a high-level, interpreted, dynamically typed, object-oriented, and general-purpose programming language. It is known for its simple and readable syntax, which makes development faster and easier.
Python has a large standard library and a vast ecosystem.
It is widely used in web development, automation, data analysis, machine learning, artificial intelligence.

-----------------------------------------------

High-level → Easy for humans to read and write.
Interpreted → Code is executed line by line without prior compilation.
Dynamically typed → No need to declare variable types explicitly; Python determines them at runtime.
Object-oriented → Supports programming using classes and objects.
General-purpose → Can be used in many domains, not limited to a specific type of application.

-----------------------------------------------

Q2. What are the advantages and disadvantages of Python?
Python's biggest strengths are its simplicity, rich ecosystem, and rapid development, while its main limitations are slower execution speed and higher memory usage compared to compiled languages.

-----------------------------------------------


Q3. What is the difference between a compiled language and an interpreted language? Where
does Python fit?

A compiled language converts the entire source code into machine code before execution, producing an executable file. This generally results in faster execution.

An interpreted language executes the source code line by line at runtime without creating a separate executable, making it easier to debug but generally slower.
